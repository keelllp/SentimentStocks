```python
import sys
import os

# Add the project root directory to Python path
project_root = r"C:\Users\PC\OneDrive\Documents\DS project"
sys.path.append(project_root)

# Now you can import from src
from src.sentiment_analysis import SentimentAnalyzer
import pandas as pd
import numpy as np

# Initialize sentiment analyzer
sentiment_analyzer = SentimentAnalyzer()

def add_sentiment_columns(twitter_file_path):
    """Add sentiment columns to Twitter data files"""
    print(f"Processing {twitter_file_path}...")
    
    # Load the Twitter data
    df = pd.read_csv(twitter_file_path)
    
    # Check if sentiment columns already exist
    if 'Positive' in df.columns and 'Neutral' in df.columns and 'Negative' in df.columns:
        print(f"Sentiment columns already exist in {twitter_file_path}")
        return df
    
    print("Adding sentiment columns...")
    
    # Apply sentiment analysis to each tweet
    sentiment_scores = []
    
    for idx, tweet in enumerate(df['Tweet']):
        if idx % 1000 == 0:  # Progress indicator
            print(f"Processing tweet {idx}/{len(df)}")
        
        # Get combined sentiment scores
        scores = sentiment_analyzer.get_combined_sentiment(tweet)
        
        # Extract the three main sentiment scores
        sentiment_scores.append({
            'Positive': scores['vader_positive'],
            'Neutral': scores['vader_neutral'],
            'Negative': scores['vader_negative']
        })
    
    # Convert to DataFrame and add to original data
    sentiment_df = pd.DataFrame(sentiment_scores)
    df = pd.concat([df, sentiment_df], axis=1)
    
    # Save the updated file
    df.to_csv(twitter_file_path, index=False)
    print(f"Updated {twitter_file_path} with sentiment columns")
    
    return df

# Use absolute paths
icici_path = os.path.join(project_root, "data", "twitter_data_new", "ICICIBANK.csv")
hdfc_path = os.path.join(project_root, "data", "twitter_data_new", "HDFCBANK.csv")

# Check if files exist
print(f"ICICIBANK exists: {os.path.exists(icici_path)}")
print(f"HDFCBANK exists: {os.path.exists(hdfc_path)}")

# Process ICICIBANK first
icici_df = add_sentiment_columns(icici_path)

# Process HDFCBANK second  
hdfc_df = add_sentiment_columns(hdfc_path)

print("Both files processed successfully!")
```

```python
import pandas as pd
import numpy as np
import os

# Load the combined data
project_root = r"C:\Users\PC\OneDrive\Documents\DS project"
processed_data_dir = os.path.join(project_root, "processed_data")

print("=== LOADING COMBINED DATA ===")

# Load combined stock data
stock_data_path = os.path.join(processed_data_dir, "combined_stock_data.csv")
stock_df = pd.read_csv(stock_data_path)
print(f"Stock data loaded - Shape: {stock_df.shape}")
print(f"Stock data columns: {list(stock_df.columns)}")

# Load combined Twitter data
twitter_data_path = os.path.join(processed_data_dir, "combined_twitter_data.csv")
twitter_df = pd.read_csv(twitter_data_path)
print(f"Twitter data loaded - Shape: {twitter_df.shape}")
print(f"Twitter data columns: {list(twitter_df.columns)}")

print("\n" + "="*50)
print("=== CLEANING DATETIME DATA ===")

# Check for problematic datetime values in Twitter data
print("Checking Twitter datetime column...")
print(f"Twitter datetime sample: {twitter_df['Datetime'].head(10)}")
print(f"Twitter datetime unique values count: {twitter_df['Datetime'].nunique()}")

# Clean Twitter datetime column - remove invalid rows
print("Cleaning invalid datetime values...")
original_twitter_count = len(twitter_df)

# Convert to datetime with error handling and localize to UTC
twitter_df['Datetime'] = pd.to_datetime(twitter_df['Datetime'], errors='coerce')
twitter_df['Datetime'] = twitter_df['Datetime'].dt.tz_localize(None)  # Remove timezone info

# Remove rows with invalid datetime
twitter_df = twitter_df.dropna(subset=['Datetime'])
cleaned_twitter_count = len(twitter_df)

print(f"Removed {original_twitter_count - cleaned_twitter_count} rows with invalid datetime")
print(f"Twitter data after cleaning - Shape: {twitter_df.shape}")

# Convert stock datetime column
print("Converting stock datetime column...")
stock_df['Datetime'] = pd.to_datetime(stock_df['Datetime'], errors='coerce')
stock_df['Datetime'] = stock_df['Datetime'].dt.tz_localize(None)  # Remove timezone info
stock_df = stock_df.dropna(subset=['Datetime'])
print(f"Stock data after cleaning - Shape: {stock_df.shape}")

print("\n" + "="*50)
print("=== PREPARING FOR MERGE ===")

# Set Datetime as index for stock data
stock_df.set_index('Datetime', inplace=True)

# Aggregate Twitter data by date and stock (daily summaries)
print("Aggregating Twitter data by date and stock...")

# Extract just the date part (without time) for grouping
twitter_df['Date'] = twitter_df['Datetime'].dt.date

twitter_daily = twitter_df.groupby(['Stock', 'Date']).agg({
    'Tweet': 'count',
    'Likes': 'sum',
    'Retweets': 'sum',
    'Replies': 'sum',
    'Quotes': 'sum'
}).rename(columns={'Tweet': 'tweet_count'})

# Add sentiment columns if they exist
if 'Positive' in twitter_df.columns:
    sentiment_cols = ['Positive', 'Neutral', 'Negative']
    for col in sentiment_cols:
        if col in twitter_df.columns:
            twitter_daily[f'{col}_mean'] = twitter_df.groupby(['Stock', 'Date'])[col].mean()
            twitter_daily[f'{col}_std'] = twitter_df.groupby(['Stock', 'Date'])[col].mean()

# Reset index to get Stock and date as columns
twitter_daily.reset_index(inplace=True)
twitter_daily['Datetime'] = pd.to_datetime(twitter_daily['Date'])
twitter_daily.set_index(['Stock', 'Datetime'], inplace=True)

print(f"Twitter daily aggregated - Shape: {twitter_daily.shape}")
print(f"Twitter daily columns: {list(twitter_daily.columns)}")

print("\n" + "="*50)
print("=== MERGING DATA ===")

# Merge stock and Twitter data
merged_dataframes = []

for stock in stock_df['Stock'].unique():
    print(f"Processing {stock}...")
    
    # Get stock data for this stock
    stock_subset = stock_df[stock_df['Stock'] == stock].copy()
    stock_subset = stock_subset.reset_index()
    stock_subset.set_index(['Stock', 'Datetime'], inplace=True)
    
    # Get Twitter data for this stock
    if stock in twitter_daily.index.get_level_values('Stock'):
        twitter_subset = twitter_daily.loc[stock]
        
        # Reset index to get Datetime as a column for easier merging
        stock_subset = stock_subset.reset_index()
        twitter_subset = twitter_subset.reset_index()
        
        # Merge stock and Twitter data on Datetime
        merged = stock_subset.merge(twitter_subset, 
                                  on='Datetime', 
                                  how='left',
                                  suffixes=('_stock', '_twitter'))
        
        # Forward fill missing Twitter data
        merged = merged.fillna(method='ffill')
        
        # Add rolling averages for Twitter metrics
        for col in ['tweet_count', 'Likes', 'Retweets', 'Replies', 'Quotes']:
            if col in merged.columns:
                merged[f'{col}_7d_avg'] = merged[col].rolling(window=7).mean()
                merged[f'{col}_30d_avg'] = merged[col].rolling(window=30).mean()
        
        merged_dataframes.append(merged)
        print(f"✅ {stock} merged - Shape: {merged.shape}")
    else:
        print(f"⚠️ No Twitter data found for {stock}")

# Combine all merged data
if merged_dataframes:
    final_merged_data = pd.concat(merged_dataframes, axis=0)
    print(f"\nFinal merged dataset shape: {final_merged_data.shape}")
    print(f"Final columns: {list(final_merged_data.columns)}")
    
    # Save merged data
    merged_output_path = os.path.join(processed_data_dir, "merged_stock_twitter_data.csv")
    final_merged_data.to_csv(merged_output_path)
    print(f"✅ Merged data saved to {merged_output_path}")
    
    # Show sample of merged data
    print(f"\nSample of merged data:")
    print(final_merged_data.head())
    
else:
    print("❌ No data was merged successfully!")
```

```python
# Analyze the merged dataset columns
print("=== ANALYZING MERGED DATASET COLUMNS ===")

# Load merged data
merged_output_path = os.path.join(processed_data_dir, "merged_stock_twitter_data.csv")
data = pd.read_csv(merged_output_path)

print(f"Dataset shape: {data.shape}")
print(f"Total columns: {len(data.columns)}")

print("\n" + "="*50)
print("=== COLUMN BREAKDOWN ===")

# Group columns by category
stock_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits']
twitter_cols = ['tweet_count', 'Likes', 'Retweets', 'Replies', 'Quotes']
sentiment_cols = ['Positive_mean', 'Positive_std', 'Neutral_mean', 'Neutral_std', 'Negative_mean', 'Negative_std']
avg_cols = [col for col in data.columns if '_avg' in col]
other_cols = ['Stock', 'Datetime', 'Date']

print("�� STOCK DATA COLUMNS:")
for col in stock_cols:
    if col in data.columns:
        null_pct = (data[col].isnull().sum() / len(data)) * 100
        print(f"  - {col}: {null_pct:.1f}% null")

print("\n🐦 TWITTER METRICS COLUMNS:")
for col in twitter_cols:
    if col in data.columns:
        null_pct = (data[col].isnull().sum() / len(data)) * 100
        print(f"  - {col}: {null_pct:.1f}% null")

print("\n😊 SENTIMENT COLUMNS:")
for col in sentiment_cols:
    if col in data.columns:
        null_pct = (data[col].isnull().sum() / len(data)) * 100
        print(f"  - {col}: {null_pct:.1f}% null")

print("\n📈 ROLLING AVERAGE COLUMNS:")
for col in avg_cols:
    if col in data.columns:
        null_pct = (data[col].isnull().sum() / len(data)) * 100
        print(f"  - {col}: {null_pct:.1f}% null")

print("\n🔧 OTHER COLUMNS:")
for col in other_cols:
    if col in data.columns:
        null_pct = (data[col].isnull().sum() / len(data)) * 100
        print(f"  - {col}: {null_pct:.1f}% null")

print("\n" + "="*50)
print("=== POTENTIALLY USELESS COLUMNS ===")

# Check for redundant columns
redundant_cols = []

# Convert Datetime to datetime first, then check if Date and Datetime are the same
if 'Date' in data.columns and 'Datetime' in data.columns:
    try:
        data['Datetime'] = pd.to_datetime(data['Datetime'])
        if data['Date'].equals(data['Datetime'].dt.date):
            redundant_cols.append('Date')
            print("⚠️ 'Date' column is redundant with 'Datetime'")
    except:
        print("⚠️ Could not compare Date and Datetime columns")

# Check for columns with all zeros
zero_cols = []
for col in data.columns:
    if col not in ['Stock', 'Datetime', 'Date']:  # Skip identifier columns
        if (data[col] == 0).all():
            zero_cols.append(col)
            print(f"⚠️ '{col}' column has all zeros")

# Check for columns with very low variance
low_var_cols = []
for col in data.columns:
    if col not in ['Stock', 'Datetime', 'Date'] and data[col].dtype in ['float64', 'int64']:
        try:
            if data[col].std() < 0.001:  # Very low standard deviation
                low_var_cols.append(col)
                print(f"⚠️ '{col}' column has very low variance (std: {data[col].std():.6f})")
        except:
            pass

print(f"\nRedundant columns: {redundant_cols}")
print(f"Zero-value columns: {zero_cols}")
print(f"Low-variance columns: {low_var_cols}")

# Show sample data
print(f"\nSample of merged data:")
print(data.head())
```

```python
# Simple cleanup - handle the few null values in rolling averages
print("=== FINAL DATA CLEANUP ===")

# Load merged data
merged_output_path = os.path.join(processed_data_dir, "merged_stock_twitter_data.csv")
data = pd.read_csv(merged_output_path)

print(f"Original shape: {data.shape}")

# Fill null values in rolling average columns (very few nulls, safe to fill)
rolling_avg_cols = [col for col in data.columns if '_avg' in col]
print(f"Rolling average columns: {rolling_avg_cols}")

# Forward fill the few null values in rolling averages
data_clean = data.fillna(method='ffill')

print(f"✅ Filled null values in rolling average columns")
print(f"Final shape: {data_clean.shape}")

# Save the final cleaned dataset
final_output_path = os.path.join(processed_data_dir, "final_cleaned_data.csv")
data_clean.to_csv(final_output_path, index=False)
print(f"✅ Final cleaned data saved to {final_output_path}")

print(f"\nDataset is ready for feature engineering and model training!")
print(f"Total features: {len(data_clean.columns)}")
print(f"Total samples: {len(data_clean)}")
```
